```python
list = []
for elem in main_data['last_price']:
    if elem <= 40000000:
        list.append(elem)
    else: list.append(None)
main_data['last_price'] = list
main_data['last_price'] = main_data['last_price'].fillna(main_data['last_price'].median())

list = []
for elem in main_data['total_area']:
    if elem <= 600:
        list.append(elem)
    else: list.append(None)
main_data['total_area'] = list
main_data['total_area'] = main_data['total_area'].fillna(main_data['total_area'].median())

columns = ['last_price', 'total_area']
for column in columns:
    plt.figure()
    main_data.boxplot([column])
```
